```python
import pandas as pd
df = pd.read_csv("movies.csv")
print(df)
```

```python
genre_choice = input("which genre you want(Sci-Fi,action.romance,comedy,drama,animation,horror ):")
#input to from user 
```

```python
filtered = df[ df["genre"] == genre_choice]
print(filtered)

#filtered data according to user command
```

```python
sorted_movies = filtered.sort_values( by="rating" , ascending= False)
print(sorted_movies)
#sorting movies according to by their rating
```

```python
top_3 = sorted_movies.head(3)
print("Top recommendations:")
print(top_3[["title","rating","year"]])

#giving top 3 highest rating movies to user as result
```

```python
def recommend_movies(genre):
    filtered = df[df["genre"] == genre]
    sorted_movies = filtered.sort_values("rating", ascending= False)
    top_3 = sorted_movies.head(3)
    return top_3[["title","rating","year",]]

genre_choice = input("kaunsa genre chahiye?")
result = recommend_movies(genre_choice)
print(result)

#combining all above program in function 
#running them at once and give one final result
```

```python
def recommend_movies(genre):
    filtered= df[df["genre"].str.lower()== genre.lower()]
    if filtered.empty:
        return "please enter  corrted keywords"
    sorted_movies = filtered.sort_values("rating", ascending=False)
    return sorted_movies.head(3)[["title","rating","year"]]

genre_choice = input("kaunsa genre hona?")
result = recommend_movies(genre_choice)
print(result)

#adding conditional statement 
#in case user enter any wrong input and lower case word
```
