```python
class car:
    def __init__ (self,brand,color):
        self.brand = brand 
        self.color =color
    
    def display(self):
        print(f"yeh ek {self.brand} hai iska color {self.color}")
car1=car("honda ","green")
car2=car("verna","white")

car1.display()
car2.display()
```

```python
#simple class
class student:
    def __init__(self,name,age,marks):
        self.name= name
        self.age= age
        self.marks= marks

    def result(self):
        if self.marks > 40:
            print(f"your pass, name : {self.name}")
        elif self.marks <= 40:
            print(f"your fail , name:{self.name}")
s1 = student("abhijeet",18,89)
s2= student("rakesh",30,90)
s3= student("ravi",20,23)

s1.result()
s2.result()
s3.result()

```

```python
class book:
    def __init__(self,tittle,author,price):
        self.tittle=tittle
        self.author=author
        self.price=price

    def showinfo(self):
        print("Tittle:",self.tittle)
        print("author:",self.author)
        print("price:",self.price)
 
s1=book("godan","munshi premchand",599)

s2= book("the writer ","abhijeet",799)
s1.showinfo()
s2.showinfo()
```

```python
class bankacc:
    def __init__(self,balance):
        self.balance = balance

    def deposit(self,amount):
        self.balance = self.balance + amount
        print(f"Deposit hua . naya balamce : {self.balance}")
    
    
    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance = self.balance-amount
            print(f"withdraw hua.naya balance: {self.balance}")

acc1 = bankacc(1000)
acc1.deposit(500)
acc1.withdraw(200)
acc1.withdraw(2000)
```
