```python
while True:
   try:
       a = float(input("number daalo:"))
       b = float(input("number daalo:"))
       op = input("Operation (+,-,*,/): ")
       
       if op == "+":
           print(a+b)
       elif op == "-":
           print(a-b)
       elif op == "*":
           print(a*b)
       elif op == "/":
           print(a/b)
       else:
           print("invalid operation")
   except ZeroDivisionError:
       print("Zero se divide nhi hoga")
   except ValueError:
       print("number input kijiye")
          
```

```python
while True:
    number =[]
    while len(number)<5:
        try:
            num= int(input("number daalo: "))
            number.append(num)
        except ValueError:
            print("yeh number nhi hai ,try again ")
    print("number : ",number)
    print("sum :",sum(number),"ab phirse")
    
    
```

```python
class student:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def show_d(self):
        print(f"Name: {self.name} aur age {self.age}")

s1=student("abhijeet",20)
s2=student("rana",84)
s1.show_d()
s2.show_d()
```
