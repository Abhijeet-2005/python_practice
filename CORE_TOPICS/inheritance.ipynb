```python
class vehical:
    def __init__(self,brand,speed):
        self.brand = brand
        self.speed = speed
    def info(self):
        print(f"the car brand is {self.brand}")
        print(f"the car top speed is {self.speed} km/h")

class car(vehical):
    def __init__(self, brand, speed,doors):
        super().__init__(brand, speed)
        self.doors=doors
#super()-yeh parent class ke methods aur ___init___ ko call karne ka tarika hai .ye inheritanceka sabse imp concept hai
    def info(self):
      super().info()
      print(f"door: {self.doors}")

class bike(vehical):
        def __init__(self, brand, speed,type):
            super().__init__(brand, speed)
        def info(self):
             super().info()
             print("type:{self.type}")
c1 = car("toyoto",200,4)
b1= bike("puslar N250",200,"sports")

c1.info()
b1.info()
        
```

```python
class employee:
    def __init__(self,name,sal):
        self.name = name
        self.sal = sal 
    def show(self):
        print(f"Employee name: {self.name} \n Salary: {self.sal}")

class manager(employee):
    def __init__(self, name, sal,depm):
        super().__init__(name, sal)   
        self.depm = depm
        
    def show(self):
        super().show()
        print(f"department:{self.depm}")

class developer(employee):
    def __init__(self, name, sal,lang):
        super().__init__(name, sal)
        self.lang= lang

    def show(self):
        super().show()
        print(f"language:{self.lang}")

m1 = manager("abhijeet",100000,"ML ")
d1 = developer("rohan",50000,"py")

m1.show()
d1.show()
```
