```python
def calculator(a,b,op):
    try:
        if op == "+":
             print(a+b)
        elif op == "-":
                print(a-b)
        elif op == "*":
                print(a*b)
        elif op == "/":
                 print(a/b)
        else:
              print("invalid opertion")


    except ValueError:
          print("you entered worng number")
    except ZeroDivisionError:
          print("can't divide by zero ")

a = int(input("enter number :"))
b = int(input("enter number :"))
op = input("opertion (+,-,*,/):")

calculator(a,b,op)

```

```python
#oop
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def greet(self):
        print(f"i'm {self.name} and age {self.age}")

# p1 = person("abhijeet",20)
# p1.greet()
        
class student(person):
    def __init__(self, name, age,college):
        super().__init__(name, age)
        self.college = college
    def greet(self):
        print(f"i'm {self.name} and age {self.age} i studying in {self.college}")

s1 = student("abhi",20,"JIT")
s1.greet()
```
