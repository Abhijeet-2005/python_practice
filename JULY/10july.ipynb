```python
class calculator:
    def add(self,a,b):
        return(a+b)
    def sub(self,a,b):
        return(a-b)
    def mul(self,a,b):
        return(a*b)
    def divide(self,a,b):
        return(a/b)
    
calc = calculator()
print(calc.add(4,5))
print(calc.divide(3,5))
print(calc.mul(8,4))
print(calc.sub(4,5))
```

```python
class shop:
    def __init__(self,item_name,price,quantity):
        self.item_name= item_name
        self.price=price
        self.quantity=quantity

    def total_cost(self):
       return self.price * self.quantity
    
    
    def apply_discount(self,percent):
        discount =(percent/100)*self.price
        new_price =self.price-discount

        print(f"new price: {new_price}")

item1= shop("apple",100,5)
print(item1.total_cost())
item1.apply_discount(10)
```

```python
class animal:
    def __init__(self,name):
        self.name= name

    def speak(self):
        print(f"{self.name} awaaz kaarta hai")

class dog(animal):
    def speak(self):
        print(f"{self.name} bolta hai - woof")

class cat(animal):
    def speak(self):
        print(f"{self.name} bolti hai - meow")

d1=dog("tommy")
c1=cat("kitty")

d1.speak()
c1.speak()
```
