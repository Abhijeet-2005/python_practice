```python
class calculator:
    def add(self,a,b):
        return(a+b)
    def sub(self,a,b):
        return(a-b)
    def mul(self,a,b):
        return(a*b)
    def divide(self,a,b):
        return(a/b)
    
calc = calculator()
print(calc.add(4,5))
print(calc.divide(3,5))
print(calc.mul(8,4))
print(calc.sub(4,5))
```
