```python
import pandas as pd 
data = {
    "Name":["Rahul","priya","amit","sneha","vikas","neha","karan","anajali"],
    "Age":[20,20,21,29,20,23,24,33],
    "Mark":[85,45,90,90,59,49,92,98],
    "City":["delhi","mumbai","pune","delhi","nagpur","nashik","raigad","bhandara"]

}
df = pd.DataFrame(data)
print(df)
```

```python
# task 1: multiple condition filter 
result = df[(df["Mark"]>70) & (df["City"] == "delhi")]
print(result)
```

```python
# task 2 : or condition 
result2 = df[(df["Mark"]>90) | (df["City"] == "delhi")]
print(result2)
```

```python
# task 3: group by (new concept)
city_avg = df.groupby("City")["Mark"].mean()
print(city_avg)
#concept groupby() - data ko category ke hisaad se group karta (jaise sab delhi wale ek sath ),phir .mean() se average nikaltahai.
```

```python
# task 4: new column add karna 
df["Result"] = df["Mark"].apply(lambda x:"Pass" if x >= 60 else "Fail")
print(df)

#.apply() - har value pe ek function chalata hai . lambda - ek chhota inline function hai (bina naam ke function).
```

```python
# task 5 :fail student ki list alga nikala (combine sab kuch )

fail_students = df[df["Result"] == "Fail"]
sorted_fail = fail_students.sort_values("Age")
final_result = sorted_fail[["Name","Mark"]]
print(final_result)
```
