```python
#simple class
class student:
    def __init__(self,name,age,marks):
        self.name= name
        self.age= age
        self.marks= marks

    def result(self):
        if marks >= 40:
            print(f"your pass, name : {self.name}")
        elif marks < 40:
            print(f"your fail , name:{self.name}")
s1 = student("abhijeet",18,89)
s2= student("rakesh",30,90)
s3= student("ravi",20,23)

s1.result()
s2.result()
s3.result()

```
