```python
import numpy as np 
matrix = np.array([[1,2,3],[4,5,6]])
print(matrix)

print( "matrix shape :",matrix.shape)

print(matrix[0])
print(matrix[0][1])
print(matrix[1][2])

```

```python

#slicing concept
arr = np.array([10,20,30,40,50])
print(arr[1:4])
print(arr[:3])
print(arr[2:])
```
